```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.table as tbl
import seaborn as sns

# Set the style
sns.set_theme(style="darkgrid", palette="muted")

# Define the data
data = {
    "Construct": [
        "Nationalism Score (Moderator)",
        "Moral Alignment (DV1)",
        "Message Efficacy (DV2)",
        "Defensive Nationalism (DV3)",
        "Policy Support (DV4)"
    ],
    "Questions Used": [
        "Q5, Q6, Q7",
        "Q8, Q9",
        "Q10",
        "Q11Q13",
        "Q15"
    ],
    "Description": [
        "composite index of scores (15). averaged (High (3), Low (<3)",
        "Captures identification with U.S. government post-criticism.",
        "Indicates if respondents perceived mutual understandin.",
        "Measures rejection or acceptance of the speaker and message.",
        "Evaluates agreement with the criticized U.S. digital policy."
    ]
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(18, 3))
ax.axis("off")

table = tbl.table(ax, cellText=df.values, colLabels=df.columns, loc="center", cellLoc='left')

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 2)
# Set header cell properties to bold and black
for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_text_props(weight='bold', color='black')
        cell.set_facecolor('lightgrey')

# Add caption
plt.figtext(0.5, 0.01, "Table 1: Variable Construction and Survey Measures", ha="center", fontsize=12, fontweight="bold", fontstyle="italic")

# Save the figure
file_path = "Variable_Construction_Table.png"
plt.savefig(file_path, bbox_inches="tight", dpi=300)

file_path
```
